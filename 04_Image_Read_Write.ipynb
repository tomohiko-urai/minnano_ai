# OpenCVによる画像ファイルの読み書き

```python
#@title インポート

import cv2
import numpy as np

from google.colab.patches import cv2_imshow
```

# 画像データを作成する

```python
size = np.array([480, 640, 3])
img = np.full(size, (223, 223, 223), dtype = np.uint8)

color = np.array([0., 255., 0.])  # BGR

cv2.line(img = img, pt1 = (0, 140), pt2 = (640, 340), color = color, thickness = 5)
cv2.circle(img = img, center = (200, 200), radius = 100, color = [255., 0., 0.], thickness = -1, lineType = cv2.LINE_AA)
cv2.circle(img = img, center = (440, 280), radius = 100, color = [255., 0., 0.], thickness = 2, lineType = cv2.LINE_AA)

# 表示する
cv2_imshow(img)
```

# 画像ファイルを書く

```python
cv2.imwrite('a.jpg', img)
```

# 画像ファイルを読む

```python
imgR = cv2.imread('a.jpg')

cv2_imshow(imgR)
```
